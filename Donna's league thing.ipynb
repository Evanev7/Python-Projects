```python
from itertools import permutations as perms
import numpy as np
```

```python
#Order is P, K, E, S, W, W again
#Groups are defined by S1, E1, K1, P1
#We don't care about E3 vs E4
Pool2 = [1,2,3,5]
Pool3 = [1,2,4,6]
Pool4 = [1,2,3,5]

#2020?
#Pool2 = [1,1,2,3]
#Pool3 = [2,3,4,5]
#Pool4 = [1,3,4,6]

#2021? Swapped P1 and P2
#Pool2 = [1,2,3,5]
#Pool3 = [3,4,1,2]
#Pool4 = [4,1,2,5]
```

```python
#We want to produce group P, K, E, S with 3 new elements, one from each pool.
sol = []
for i in perms(Pool2):
    for j in perms(Pool3):
        for k in perms(Pool4):
            Group = [0,0,0,0]
            for l in range(4):
                x = [l+1,i[l],j[l],k[l]]
                if len(set(x)) < len(x):
                    break
                Group[l] = x
            else:
                sol.append(Group)
```

```python
letters = ["P", "K", "E", "S", "W", "W"]

for i in range(len(sol)):
    print("Iteration "+str(i))
    for j in range(4):
        print("Group "+letters[j] +": " + " ".join([letters[sol[i][j][k]-1]+str(k+1) for k in range(4)]))
    print("")
```

```python
flatsol = []
names = []
for i in range(len(sol)):
    bwamp = []
    for j in range(4):
        for k in range(4):
            if sol[i][k][j]+6*j not in names:
                names.append(sol[i][k][j]+6*j)
            bwamp.append(names.index(sol[i][k][j] + 6*j))
    flatsol.append(bwamp)
    
for i in range(1,len(flatsol)):
    used = []
    cycles = []
    index = 0
    while index <= 15:
        if index == flatsol[i][index]:
            used.append(index)
            index += 1
        elif index in used:
            index += 1
        else:
            cycle = [index]
            current = flatsol[i][index]
            while current != index:
                cycle.append(current)
                current = flatsol[i][current]
            for thing in cycle:
                used.append(thing)
            cycles.append(cycle)
    print(max([len(i) for i in cycles]), cycles)
```
