```python
from scipy.stats import *
from matplotlib import pyplot as plt
from random import randint
```

```python
r = 20*skewnorm.rvs(5,size=10000)+21
```

```python
fig = plt.figure(figsize=[20,20])
ax = fig.gca()
c=(min(r))//1
cs = c
d = [0]
for i in sorted(r):
    d[-1] += 1
    if i > c:
        c+=1
        d.append(0)
ax.bar(range(int(cs),int(c)+1),d)
plt.show()
plt.close()
```

```python
def meems():
    meem = 0
    for i in range(10):
        meem += (r[randint(0,9999)])/10
    return meem
```

```python
datatoo = []
for i in range(10000):
    datatoo.append(meems())
fig = plt.figure(figsize = [20,20])
ax = fig.gca()
c=(min(datatoo))//1
cs = c
d = [0]
for i in sorted(datatoo):
    d[-1] += 1
    if i > c:
        c+=1
        d.append(0)
ax.bar(range(int(cs),int(c)+1),d)
plt.show()
plt.close()
```
