```python
def isprime(n):
    for i in range(2,round((n)**.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=[20,20])
ax = fig.gca()

loc = [["" for _ in range(100)] for _ in range(100)]
coord = [50,50]
dire = "down"
n=1
try:
    while True:
        print(n,isprime(n),coord)
        if isprime(n):
            loc[coord[0]][coord[1]] = True
        else:
            loc[coord[0]][coord[1]] = False
        n+=1
        if dire == "down":
            if loc[coord[0]][coord[1]-1] == "":
                dire = "left"
                coord[1] -= 1
            else:
                coord[0] += 1
        elif dire == "left":
            if loc[coord[0]-1][coord[1]] == "":
                dire = "uppp"
                coord[0] -= 1
            else:
                coord[1] -= 1
        elif dire == "uppp":
            if loc[coord[0]][coord[1]+1] == "":
                dire = "rigt"
                coord[1] += 1
            else:
                coord[0] -= 1
        else:
            if loc[coord[0]+1][coord[1]] == "":
                dire = "down"
                coord[0] += 1
            else:
                coord[1] += 1
except Exception as e:
    print(e)
print(loc)
for i in range(len(loc)):
    for j in range(len(loc[i])):
        if loc[i][j] == True:
            ax.scatter(i,j, s=200)
    
plt.show()
plt.close(fig)
```
