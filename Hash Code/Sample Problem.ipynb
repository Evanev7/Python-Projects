```python
def without(subData, data):
    return [i for i in data if i not in subData]
```

```python
dataSet = "d"
```

```python
with open(dataSet + ".in") as file:
    data = file.read().split("\n")[0:2]
    goal, total = list(map(int, data[0].split(" ")))
    data = list(map(int, data[1].split(" ")))
```

```python
data.insert(0,0)
subData = data[1:]
```

```python
def smallestChange(subData, data, goal, v=False, kill = False):
    opts = without(subData,data)
    t = len(opts)
    maxError = abs(sum(subData) - goal)
    best = []
    for i in range(len(subData)):
        temp = subData[i]
        for j in range(t):
            subData[i] = opts[j]
            error = abs(sum(subData) - goal)
            if error < 100:
                if v:
                    print("error:\n", error)
                #maxError = error
                best = list(subData)
    
        subData[i] = temp
    return best
```

```python
while abs(sum(subData) - goal) > 10 and sum(subData) != 0:
    t = smallestChange(subData, data, goal)
    if sum(t) == 0:
        break
    else:
        subData = t
    print(sum(subData) - goal)
    if 0 in subData:
        subData.remove(0)
```
