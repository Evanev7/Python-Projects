```python
from random import randint
from sympy import prime
```

```python
n = 32
primes = [prime(i) % n for i in range(2,n+2)]
players = [set([i]) for i in range(n)]
for i in range(len(players)):
    rand = 0
    print(rand, primes[i])
    while len(players[i]) < n//2:
        rand = (rand + primes[i]) % n
        if i == rand:
            continue
        if len(players[rand]) == n//2:
            continue
        players[i].add(rand)
        players[rand].add(i)
    print(players[i])
```

```python
def detect_cycle(players):
    triads = []
    for i, team_i in enumerate(players):
        for player_j in team_i:
            if player_j == i:
                continue
            for player_k in players[player_j]:
                if player_k == player_j:
                    continue
                if player_k == i:
                    continue
                if player_k in team_i:
                    triads.append((i, player_j, player_k))
        
    return triads

triads = detect_cycle(players)
len(triads)
```

```python
import networkx
from math import cos, sin, pi
```

```python
def conv_to_graph(players):
    out = []
    for i in range(len(players)):
        for j in players[i]:
            if i == j:
                continue
            if (i,j) in out:
                continue
            if (j,i) in out:
                continue
            
            out.append((i,j))
    return out

graph_players = conv_to_graph(players)
```

```python
graph_players
```

```python
g = networkx.Graph()
g.add_nodes_from(range(n))
for edge in graph_players:
    g.add_edge(*edge)
```
