```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def iterate(state, t, dt, consts=[]):
    fig = plt.figure(figsize=[15,15])
    ax=fig.gca()
    data = [["Time","Displacement","Velocity","Accelelration"],[0]+state]
    for i in range(1,int(t/dt)+1):
        data.append([i*dt, 
                     data[-1][1] + data[-1][2]*dt, 
                     data[-1][2] + data[-1][3]*dt, 
                     -(consts[0]**2)*data[-1][1]])
    
    ax.scatter([float(i) for i in np.array(data)[1:, 0]],[float(i) for i in list(np.array(data)[1:, 1])])
```

```python
iterate([0,20,0],30,.001,[1])
```
